## **Ejercicio 3**

Ahora que ya tienes el dataset cargado y listo para ser utilizado, escoge al menos 3 algoritmos de los vistos en la teoría/notebooks de la asignatura y aplícalos para clasificarlo. Una vez obtenidos los resultados, compáralos y explica tus conclusiones teniendo en cuenta lo que pensabas conseguir al elegir dicho dataset.

Normalmente, es buena costumbre aplicar un *k-fold cross-validation* para evaluar un algoritmo, para de esta forma eliminar al máximo posible el sesgo introducido al hacer la separación de datos en *train* y *test*. Para ello, puedes utilizar la siguiente instrucción que, además, realiza el *k-fold* `n_repeats` veces con distintas semillas, para mitigar todavía más dicho sesgo:

```
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=1)
scores = cross_val_score(model, X, y, scoring='accuracy', cv=cv, n_jobs=-1, error_score='raise')
```

Este código realizará el *k-fold cross-validation* con `k=10`, y lo repetirá 3 veces, una con `random_state=1`, la siguiente con `random_state=2` y la última con `random_state=3`.



````python
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
from sklearn.model_selection import train_test_split

seed = 42

# dividimos en inputs y outputs
values = dataset.values
X = values[:,0:8]
y = values[:,8]

# estandarizamos los datos
sc = StandardScaler()
X_scaled = sc.fit_transform(X)
````

Vamos a utilizar los algoritmos: Decision Tree, SVM y Perceptrón para clasificar nuestros datos y compararlos.

**DecisionTree**

````python
from sklearn.tree import DecisionTreeClassifier

model = DecisionTreeClassifier(random_state=seed)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=seed)
scores = cross_val_score(model, X_scaled, y, scoring='f1', cv=cv, n_jobs=-1, error_score='raise')

# report the mean performance
print('F1-score: %.3f' % scores.mean())

# visualizamos la matriz de confusión para una ejecución en concreto
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=0)
model.fit(X_scaled_train, y_train)
y_pred = model.predict(X_scaled_test)
cm = metrics.ConfusionMatrixDisplay(confusion_matrix=metrics.confusion_matrix(y_pred, y_test, labels=model.classes_), display_labels=model.classes_)
cm.plot(cmap="magma")
````

**SVM**

````python
from sklearn import svm

model = svm.SVC(random_state=seed)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=seed)
scores = cross_val_score(model, X_scaled, y, scoring='f1', cv=cv, n_jobs=-1, error_score='raise')

# report the mean performance
print('F1-score: %.3f' % scores.mean())

# visualizamos la matriz de confusión para una ejecución en concreto
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=0)
model.fit(X_scaled_train, y_train)
y_pred = model.predict(X_scaled_test)
cm = metrics.ConfusionMatrixDisplay(confusion_matrix=metrics.confusion_matrix(y_pred, y_test, labels=model.classes_), display_labels=model.classes_)
cm.plot(cmap="magma")
````

**SVM (Radial Basis Function kernel)**

````python
from sklearn import svm

model = svm.SVC(kernel='poly', random_state=seed)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=seed)
scores = cross_val_score(model, X_scaled, y, scoring='f1', cv=cv, n_jobs=-1, error_score='raise')

# report the mean performance
print('F1-score: %.3f' % scores.mean())

# visualizamos la matriz de confusión para una ejecución en concreto
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=0)
model.fit(X_scaled_train, y_train)
y_pred = model.predict(X_scaled_test)
cm = metrics.ConfusionMatrixDisplay(confusion_matrix=metrics.confusion_matrix(y_pred, y_test, labels=model.classes_), display_labels=model.classes_)
cm.plot(cmap="magma")
````

**Perceptrón**

````python
from sklearn.linear_model import Perceptron

model = Perceptron(random_state=seed)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=seed)
scores = cross_val_score(model, X_scaled, y, scoring='f1', cv=cv, n_jobs=-1, error_score='raise')

# report the mean performance
print('F1-score: %.3f' % scores.mean())

# visualizamos la matriz de confusión para una ejecución en concreto
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=0)
model.fit(X_scaled_train, y_train)
y_pred = model.predict(X_scaled_test)
cm = metrics.ConfusionMatrixDisplay(confusion_matrix=metrics.confusion_matrix(y_pred, y_test, labels=model.classes_), display_labels=model.classes_)
cm.plot(cmap="magma")
````

**Logistic regression**

````python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(random_state=seed)
cv = RepeatedStratifiedKFold(n_splits=10, n_repeats=3, random_state=seed)
scores = cross_val_score(model, X_scaled, y, scoring='f1', cv=cv, n_jobs=-1, error_score='raise')

# report the mean performance
print('F1-score: %.3f' % scores.mean())

# visualizamos la matriz de confusión para una ejecución en concreto
X_scaled_train, X_scaled_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=0)
model.fit(X_scaled_train, y_train)
y_pred = model.predict(X_scaled_test)
cm = metrics.ConfusionMatrixDisplay(confusion_matrix=metrics.confusion_matrix(y_pred, y_test, labels=model.classes_), display_labels=model.classes_)
cm.plot(cmap="magma")
````
